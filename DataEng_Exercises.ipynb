```python
# Install PySpark (Kaggle may need this)
!pip install pyspark -q

from pyspark.sql import SparkSession
from pyspark.sql.functions import lit

# Start Spark
spark = SparkSession. builder. appName("DataEngineeringProject").getOrCreate()

# ===== DataFrame 1: Basic data =====
data1 = [("Alice", 25, 50000), ("Bob", 30, 60000), ("Charlie", 35, 70000), ("Diana", 28, 55000)]
df1 = spark.createDataFrame(data1, ["Name", "Age", "Salary"])
print("===== DataFrame 1 =====")
df1.show()
df1.summary().show()

# ===== DataFrame 2: Add column using lit =====
df2 = df1.withColumn("Department", lit("Engineering"))
print("===== DataFrame 2 =====")
df2.show()
df2.summary().show()

# ===== DataFrame 3: Filtered data =====
df3 = df1.filter(df1["Age"] > 28)
print("===== DataFrame 3 (filtered Age > 28) =====")
df3.show()
df3.summary().show()

# Stop Spark
spark.stop()
print("===== COMPLETED =====")
```
